```python
import numpy as np
from matplotlib import pyplot as plt
import timeit
import cProfile
import pstats
```

```python
from using_kde.precision_recall import plot_precision
```

```python
n = 400
y_act = np.linspace(0, 1, num=n)
y_pred = y_act + np.random.normal(0.0, 0.1, size=n)
```

```python
%timeit plot_precision(y_act, y_pred)
```

```python
%timeit plot_precision(y_act, y_pred)
```

```python
with cProfile.Profile() as prof:
    plot_precision(y_act, y_pred)

stats = pstats.Stats(prof)
stats.sort_stats(pstats.SortKey.TIME)
stats.print_stats()
stats.dump_stats('stats_4.prof')
```
