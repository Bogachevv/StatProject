```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

```python
from precision_recall import plot_precision, plot_recall
```

```python
size = 50
y_act = np.linspace(0, 1, num=size)
y_pred = np.linspace(0, 1, num=size) + stats.norm.rvs(loc=0.0, scale=0.01, size=size)
```

```python
plot_precision(y_pred, y_act)
```

```python
plot_precision(y_pred, y_act, plotter_mode='scatter')
```

```python
plot_recall(y_pred, y_act)
```

```python
plot_recall(y_pred, y_act, plotter_mode='scatter')
```
